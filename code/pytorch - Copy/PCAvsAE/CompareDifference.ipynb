```python
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from torchvision import transforms as tfs
from torchvision.utils import save_image
from PIL import Image
from torch import nn
from torch.autograd import Variable
from torchvision.transforms.functional import *
import os
import numpy
import torch 
import gdal


    
def minmaxnormization(tensor):
    tensor=tensor-tensor.min()
    tensor=tensor/tensor.max()
    return tensor

os.chdir(r'E:\Moore\pyScripts\pytorch\PCAvsAE')
band_fn='./imgfolder/imgraw/windowedLC08B5.rst'
in_ds = gdal.Open(band_fn)
in_band=in_ds.GetRasterBand(1)
band_raw=torch.from_numpy(in_band.ReadAsArray().astype('float'))
band_raw=minmaxnormization(band_raw)

with open(r'./output_denoised/test writer.txt','w') as f:
    for i in range(1):
        band_fn='./output_denoised/AE_1750_test10/AE_1750_rst/out_encoded{}_b1.rst'.format(i+1)
        in_ds = gdal.Open(band_fn)
        in_band=in_ds.GetRasterBand(1)
        band=torch.from_numpy(in_band.ReadAsArray().astype('float'))*(-1.0)
        band=minmaxnormization(band)
        loss=nn.MSELoss(reduction='sum')(band,band_raw).item()
        print('encoded_img{} loss: \t{:.4f}'.format(i+1,loss))
        f.write('encoded_img{} loss: \t{:.4f}\n'.format(i+1,loss))
    f.write('\n')
    for i in range(64):
        band_fn='./output_denoised/AE_1750_test10/AE_1750_rst/out_denoised{}_b1.rst'.format(i+1)
        in_ds = gdal.Open(band_fn)
        in_band=in_ds.GetRasterBand(1)
        band=torch.from_numpy(in_band.ReadAsArray().astype('float'))
        band=minmaxnormization(band)
        loss=nn.MSELoss(reduction='sum')(band,band_raw).item()
        print('denoised_img{} loss: \t{:.4f}'.format(i+1,loss))
        f.write('denoised_img{} loss: \t{:.4f}\n'.format(i+1,loss))
    
    


# band_fn='./output_denoised/PCA_1750/pca__T-Mode_Cmp1.RST'
# in_ds = gdal.Open(band_fn)
# in_band=in_ds.GetRasterBand(1)
# band=torch.from_numpy(in_band.ReadAsArray().astype('float'))
# band=minmaxnormization(band)
# loss=nn.MSELoss(reduction='sum')(band,band_raw)
# print(loss)

# with open(r'./output_denoised/test writer.txt','w') as f:
#     for i in range(64):
#         band_fn='./output_denoised/PCA_1750/imgnoised_{}_b1.rst'.format(i+1)
#         in_ds = gdal.Open(band_fn)
#         in_band=in_ds.GetRasterBand(1)
#         band=torch.from_numpy(in_band.ReadAsArray().astype('float'))*(-1.0)
#         band=minmaxnormization(band)
#         loss=nn.MSELoss(reduction='sum')(band,band_raw).item()
#         print('imgnoised{} loss: \t{:.4f}'.format(i+1,loss))
#         f.write('imgnoised{} loss: \t{:.4f}\n'.format(i+1,loss))
#     f.write('\n')


print('done')
```
